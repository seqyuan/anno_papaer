# 高温胁迫下水稻根单细胞染色质可及性的动态改变

2022年12月中国农业科学院生物技术研究所团队绘制了水稻首个单细胞染色质可及性图谱，阐述了高温胁迫下的特异细胞类群的染色质可及性动态变化，为研究作物发育和环境适应性的准确调控提供了新途径。

**文章题目**：Chromatin accessibility illuminates single-cell regulatory dynamics of rice root tips

**期刊**：BMC Biology


### 研究背景
在真核生物中，基因表达与开放染色质高度相关，可以通过高通量测序技术，分析转座酶可及染色质（ATAC-seq）进行定位。水稻是世界上最重要的经济作物之一，研究团队前期研究分析了两个品种的水稻根尖单细胞转录组，揭示了物种特有的表达程序。为阐明已研究基因表达状态的调控机制，研究者通过在单细胞分辨率观察不同组织和细胞类型的不同染色质模式，以揭示关键的调控序列。

### 材料选择

正常和热激条件下粳稻群体品种46758个根尖细胞核

### 研究结果
为了全面描述水稻根系发育过程中的染色质可及性，研究者应用单细胞ATAC-seq（scATAC-seq）对正常和热胁迫条件下的水稻栽培亚种粳稻日本晴（Nipponbare, geng）根尖的46,758个细胞进行研究，共鉴定出9种细胞类型。通过细胞类型特异性可及性染色质区域(ACRs)注释和聚类，发现并不是所有的簇特异性ACRs都定位于细胞类型标记基因。通过RNA原位杂交，发现LOC_Os04g53640的mRNA在根帽中表达与其在UMAP注释的根冠细胞群中的富集表达一致，表明细胞类群特异性ARCs有能力检测出单细胞RNA-seq无法识别的差异表达基因。随后，研究者将scATAC-seq与最近报道的水稻根尖原生质体的scRNA-seq数据进行了比较，并绘制了UMAP图，发现匹配的细胞类型之间标记基因的相关性远高于不同细胞类型之间的相关性，表明染色质可及性和基因表达数据集呈正相关，进一步证实了研究者的聚类注释结果。

### 总结

该研究报道了在单细胞分辨率下的水稻根系发育过程中的染色质景观，定义了细胞类型发育和对热应激的响应过程中的染色质动力学，对挖掘新的基因及调控网络，以及智能育种具有重要意义。
