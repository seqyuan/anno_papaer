# 自体细胞疗法修复乳腺结构和功能

文章名称：Restoring mammary gland structures and functions with autogenous cell therapy

安诺优达为本研究提供10x单细胞转录组测序及数据分析服务，安诺优达高级分析工程师为此次发表文章的合作作者。

* **研究物种**：山羊
* **发表杂志**：Biomaterials（IF：12.47）
* **文章单位**：广西大学亚热带农业生物资源保护与利用国家重点实验室
* **发表时间**：2021.09

## 主要内容
本研究使用具有山羊有泌乳功能乳腺上皮细胞（CiMEC），其具有来自山羊耳成纤维细胞（GEF）的泌乳功能，移植的CiMECs在小鼠体内再生了具有泌乳功能的正常乳腺结构。单细胞RNA测序显示，GEFs可绕过全能干细胞阶段实现泌乳功能。本研究提出了一种针对乳腺缺陷的自体细胞疗法和转基因乳腺生物反应器生产的创新策略。
