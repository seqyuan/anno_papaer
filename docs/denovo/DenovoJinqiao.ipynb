# 荞麦基因组的比较分析揭示了代谢组差异和生态型分化的遗传基础

2022年6月，中国农业科学院作物科学研究所周美亮研究员团队在国际期刊New Phytologist上发表论文。研究使用比较基因组方法，为苦荞的药理遗传性状改良和杂交育种奠定了理论基础。安诺优达为本研究提供基因组分析服务。 

<div align=center></div>

本研究使用比较基因组方法，从染色体水平揭示了金荞和苦荞遗传差异，及类黄酮物质合成相关基因的进化差异，并基于重测序划分了金荞的生态型，进一步验证了控制类黄酮物质合成的关键基因。
> 周美亮团队博士后何铭（现中国农业科学院蔬菜花卉研究所助理研究员）、博士后何毓琦、副研究员张凯旋、博士生卢翔为该论文共同第一作者，周美亮研究员为通讯作者，安诺优达为本研究提供基因组分析服务，同时也是文章共同第一作者。

## 研究背景
1）金荞富含类黄酮，一种在治疗感染、癌症和糖尿病里备受关注的物质；

2）金荞被认为是改良栽培荞麦的理想野生近缘种，通过杂交可以改良荞麦的性状。

## 研究内容
1）通过金荞和荞麦的比较基因组分析，探讨类黄酮合成的调控机制；

2）通过对34份野生材料的重测序，进一步明确潜在重要性状的候选基因；

3）希望通过基因组分析为荞麦类黄酮合成基因的进化提供了新的思路，并有望为金荞和其他植物的生物学问题研究和植物药理遗传改良提供帮助。

## 主要研究方法&结果

### 1）基因组组装
本研究使用三代测序技术和Hi-C技术构建了染色体水平的金荞基因组，大小为1.08Gb，杂合度1.02%。由于高杂合度，初步组装基因组大小为1.59Gb，去冗余之后基因组大小为1.08Gb，与流式和survey结果接近。基因组contig N50 2.84Mb，BUSCO 92.9%。0.74Gb（68.21%）基因组序列由重复序列组成，LTR重复是最常见的重复，其中Gypsy和Copia丰度最高。使用RNA、ab initio预测和同源预测方法注释到38919个基因。



### 2） 染色体结构差异
共线性分析表明，金荞和苦荞在全染色体水平上检测到1对1的共线关系，包括114个区块，共线基因20676个。苦荞自分化以来，出现了更多的缺失，这与金荞基因组大于苦荞一致。同时观测到2个大的染色体倒位Chr3 (29.5 Mb)和Chr4 (23.8 Mb)，分别包括1673个基因和759个基因。

GO富集分析表明Chr3断点附近的基因在“茎尖分生组织发育”、“碳水化合物代谢过程”和“代谢过程”显著富集，包括9种类黄酮合成相关基因。Chr4断点附近的基因与细胞壁生物合成相关通路显著富集。
### 3）Ks/Ka分析
进化分析发现，金荞和苦荞约1 Mya前分化，金荞基因组约是苦荞（0.48Gb）的两倍，但是基因数量类似，这主要是由于重复序列扩增导致。Ks/Ka分析表明物种形成事件的Ks峰对应的基因在代谢和发育过程中显著富集，包括脂质代谢、类黄酮代谢、果实发育和侧根形成，表明金荞代谢相关基因受到强烈选择。
### 4）全基因组复制分析
全基因组复制事件也驱动了金荞基因组扩张，在金荞麦和苦荞中只发现了一个Ks峰，Ks = ~0.99。这表明在物种形成之前的71.1 Mya左右发生了一次古老的全基因组复制事件（WGDs），没有最近的WGDs事件。


进一步分析表明，在金荞麦和苦荞基因组中，重复类型的分布相似，主要有散在重复、WGD和转座子重复，但金荞麦的散在重复类型占比更高。通过散在重复、近端重复和串联重复引起的基因重复在类黄酮合成基因附近更高。串联重复也在细胞壁合成相关基因附近富集。转座子重复与蛋白质磷酸化修饰和生物和非生物胁迫反应相关。WGD基因主要富集于非生物胁迫和生物胁迫及发育相关方面。



### 5）基因家族分析
石竹目（金荞、荞麦、Beta vulgaris和B. patula）的基因家族分析表明，石竹目共有10338个共享基因家族和7277个金荞和苦荞的同源基因家族。通过对金荞麦和苦荞的比较，发现了8个金荞麦特有的植物自交不亲和蛋白S1，与金荞麦的自交不亲和一致。



构建了13个物种共有344个单拷贝基因的系统发育树，结果表明，荞麦和甜菜聚在Brassicales和Rosales之外的一个分支上，与其他双子叶植物的差异约为138 Mya。



基因家族的进化在植物表型多样性中起着重要的作用。金荞鉴定出了739个扩增基因家族和528个收缩基因家族，扩增基因家族在细胞调控、合成和代谢过程富集。对鉴定到的85个细胞色素P450进行分析，涉及次级代谢和激素代谢，其中54个具有类黄酮3'-单加氧酶结构域，12个与苦荞FtF3'5'H相关。
### 6）转录&代谢产物差异分析
与苦荞相比，金荞中FAR基因拷贝数增加，这与植物生长发育、代谢过程相关。FdFRS1基因在金荞的根毛处超表达可以提高一种类黄酮（芦丁）的含量。



对金荞麦的地下膨大茎、根和苦荞的根的代谢组学分析发现，金荞中儿茶素、原花青素、芦丁、槲皮素等黄酮类物质含量远高于苦荞，这是金荞麦可做药用的重要基础。验证了基因组层面金荞中黄酮类物质合成相关基因扩增。



### 7）34份重测序数据分析
对34份直立和斜生生态型的二倍体野生金荞测序，并进行了性状评价，发现直立生态型的农艺性状优于斜生生态型，而斜生生态型的品质性状更优。进而挖掘到一系列与品质和农艺性状相关的基因，并开发了分子标记。

## 结论
综上所述，本研究提供的基因组序列为阐明金荞药用相关分子的合成及其调控机制奠定了基础，为荞麦等含有增值分子药用植物的育种或基因工程提供了参考。通过野生种的重测序结果发现金荞包含广泛的遗传变异，可为研究金荞群体结构、为今后更好地利用金荞自然遗传变异资源进行荞麦育种提供依据。
