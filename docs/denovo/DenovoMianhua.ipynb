# 高质量棉花基因组揭示陆地棉起源和驯化的新视角

浙江大学农业与生物技术学院作物科学研究所的研究人员在Journal of Advanced Research（IF=12.822）上发表了一篇关于海南岛本地棉花（Hainan Island Cotton，HIC）的研究论文，题为“Gossypium purpurascens genome provides insight into the origin and domestication of upland cotton”。该文从基因组水平解析了HIC与其他四倍体棉花的进化关系和基因组多样性，以及结构变异（Structural Variation, SV）在陆地棉驯化中的作用。这项研究不仅为棉花的起源和驯化提供了新的视野和证据，而且为棉花育种和品种改良提供重要参考依据。  
**安诺优达为本项目提供三代PacBio HiFi测序服务，同时参与项目设计和生信分析，安诺生信分析工程师张雪梅为共同第一作者，杨伟飞为共同作者。**



## 01 研究背景
陆地棉（Gossypium hirsutum L.）是世界上最重要的棉花物种之一，其产量占全球棉花总产量的90%以上。该物种起源于约100万至150万年前的美洲地区，由两个野生的二倍体棉花物种在自然界中杂交产生异源四倍体染色体陆地棉原始种群。从18世纪中期开始，陆地棉被成功驯化改良并广泛传播到世界各地。作为重要的经济作物，棉花对全球经济和社会发展产生了深远的影响。

在本研究中，研究团队聚焦于在我国海南、广东、广西等地自然生长的本地棉种——海南岛本地棉花（Hainan Island Cotton，HIC），组装了该棉种的高质量基因组，解析了其起源和进化关系，并分析了棉花大范围的关键结构变异在驯化中的作用。这些研究结果对于作物驯化和遗传改良具有重要的理论参考意义。
## 02 研究材料

海南岛本地棉花（Hainan Island Cotton，HIC）——HPF17

## 03 研究结果

基于PacBio三代HiFi测序和Hi-C辅助组装技术，研究人员组装了一个高质量的四倍体野生型棉花HPF17的基因组，基因组大小约为2.27 Gb，Contig N50为13.20 Mb。对该基因组进行注释分析，研究人员预测到79,146个编码基因，平均长度3,065 bp；鉴定到来自69个基因家族的5,558个转录因子，占蛋白质编码基因的7.02%；确定了48,491个核糖体RNA（rRNA）基因、4,854个转移RNA（tRNA）基因、823个微小RNA（miRNA）基因和12,903个小核RNA（snRNA）基因；发现了约64.67%的重复序列，主要包括0.47%的串联重复序列和58.50%的转座子序列。最后，基于12个陆地棉基因组的比较基因组分析，研究人员确定了三种变异类型，即存在/缺失变异（PAV）、倒位（INV）和易位（TRANS），平均每个基因组中鉴定出171,231个SV。



通过对HPF17和其他11个四倍体棉花的SV分析，结合SNP和QTL信息，分析遗传特征和性状差异，研究人员发现相较于非SV区域，SV区域含有更多的QTL，且这些SV区域中含有更多的选择性消除区域，INV和TRANS中选择性消除区域的占比分别为55.5%和24.5%。因此，推测SV，特别是INV和TRANS可能对陆地棉的驯化和改良发挥了重要作用。

此外，对大量陆地棉栽培种和野生种系的基因分型表明这些大范围INV可能在陆地棉驯化的早期阶段就经历了人工选择。为了进一步揭示SV对驯化的贡献，研究人员对（TM-1 × HPF17）F2群体进行基因分型，分析表型差异和INV之间的关系，发现可能由于INV的选择，现代棉花品种的种子大小、衣分、纤维质量及其他性状得到了显著改善，这表明倒位在驯化和分化中发挥重要作用。

  

为了研究陆地棉的进化历程，基于不同棉花品种的SNP分布规律，研究人员进行了系统发育分析、分歧时间估计、遗传多样性分析和种群分化分析，分析表明所有的陆地棉可以分为三组，陆地棉栽培种cultivars of G. hirsutum （GHC），蓬蓬棉G. purpurascens from South China（GHP），以及中美洲野生种系G. hirsutum landraces/races（GHL） from Central America。通过最大似然法（ML）构建的系统发育分析表明，研究对象HIC属于G. purpurascens，G. purpurascens比G. yucatatum以外的中美洲种群更原始，G. purpurascens和G. hirsutum栽培种之间的关系相对较远。综合多种分析结果，G. purpurascens被归类为G. hirsutum的一个原始种系。 

  

研究发现HIC种子具有极强的耐盐性，可以在盐水里存活半年以上，并且蓬蓬棉及其他野生棉花物种/种系在太平洋岛屿上的分布与太平洋洋流吻合度高，通过分歧时间估计发现大约20万年前海南岛本土棉花蓬蓬棉就从原产美洲的四倍体陆地棉中分化出来。因此，研究人员推测，HIC是随着太平洋洋流漂散，从中美洲西海岸逐渐西进到达中国南海诸岛。研究人员结合多种历史资料推测，早在哥伦布时期以前，美洲棉驯化栽培前，蓬蓬棉就已经在海南岛被初步驯化和小规模栽培，并用于纺织“崖州布”（世界非物质文化遗产）。因此，三亚（古称崖州）可能是陆地棉最早驯化和栽培的地点之一，即中国是四倍体棉花最早驯化栽培的国家之一。  



## 04 研究总结

该文章报道了一个基于PacBio HiFi测序和Hi-C辅助组装技术组装的高质量四倍体野生型棉花基因组，并对其进行了注释和分析。研究人员发现该基因组共有79,146个编码基因，其中约7%是转录因子，另外还富含多种RNA基因和重复序列。研究人员通过对基因组的比较分析，发现陆地棉的驯化和改良可能受到了基因组结构变异（PAV、INV和TRANS）的影响，这些变异可能对一些重要性状（如种子大小、衣分和纤维质量）的改善发挥了重要作用。此外，研究人员还进行了系统发育分析和种群分化分析，发现所有的陆地棉可以分为三组（GHC、GHP、GHL），其中G. purpurascens被归类为G. hirsutum的一个原始种。最后研究人员结合太平洋洋流分布和历史资料，揭示了中国在陆地棉驯化和栽培中的悠久历史地位。该研究对于深入了解陆地棉的基因组组成、遗传多样性和进化历程具有重要意义，也为陆地棉的驯化和育种提供了重要参考。
